```python
import pandas as pd
from google.colab import drive

drive.mount('/drive')

df = pd.read_csv('/drive/MyDrive/class_titanic.csv')
df = pd.DataFrame(df)

display(df)
```

```python
'''questao 3'''
df=df.dropna(subset=['Age'])
display(df)
```

```python
'''questao 4'''
diMedia=df['Age'].mean()
diMedia
```

```python
'''questao 5'''
#criterio: se a idade diferir em 30% ou mais da media
df.drop(df[df['Age'] > 1.3*diMedia].index, inplace = True)
df.drop(df[df['Age'] < 0.7*diMedia].index, inplace = True)
display(df)
```

```python
'''questao 6'''
df.drop(columns=['PassengerId','Ticket','Parch','Survived','Pclass','Cabin','Sex'], inplace=True)
df
```

```python
'''questao 7'''
#coluna a será a coluna "fare" e a coluna B será a coluna "Age"
col1=df['Name']
col2=['Age']
col3=df['SibSp']
col4=df['Fare']
col5=df['Embarked']

df['colunaX']=(col4-col3)*0.3
display(df)
```

```python
'''questao 8'''
import numpy as np
df=df.reindex(np.random.permutation(df.index))
pd.set_option("display.max_columns",5)
pd.set_option("display.max_rows",6)
display(df)
```

```python
'''questao 9(ordenando por ordem de idade, do menor ao maior)'''

df = df.sort_values(by='Age', ascending=True)
print(f"A pessoa mais nova é: {df['Age'].iloc[0]}")
      
pd.set_option('display.max_columns', 5)
pd.set_option('display.max_rows', 6)

display(df)
```

```python
'''questao 10'''
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 5)


grupo = df.groupby('Embarked')[['Age']].mean()

display(grupo)

#Agrupando os dados pelos totais da coluna

grupo2 = df.groupby('Name')[['Name']].count()

display(grupo2)
```

```python
'''questao 11
Criterio=criei uma coluna que diz de qual cidade os passageiros embarcaram: Southampton, Cherbourg ou Queenstown (S,C,Q)'''
df['Embarcou De'] = df['Embarked'].map(
    {'S':'Southampton','C':'Cherbourg','Q':'Queenstown'})


pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 10)
display(df)
```
